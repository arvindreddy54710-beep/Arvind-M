```python
# 10: Create a list of prime numbers between 1 and 50
```

```python
pri = []

for num in range(2, 50):
    is_pri = True

    for i in range(2, num):
        if num % i == 0:
            is_pri = False
            break

    if is_pri:
        pri.append(num)

print(pri)
```

```python
# 15: Transpose the matrix [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
```

```python
matrix =  [[1, 2, 3],
           [4, 5, 6],
           [7, 8, 9]]
tra =[]
for i in range(len(matrix[0])):
    row =[]
    for j in range(len(matrix)):
        row.append(matrix[j][i])
    tra.append(row)
print(tra)
```
