```python
# Question 3: Get Array Shape (Easy)Find the shape, size, and dimensions of the retail data array.Hint: Use .shape, .size, .ndim
```

```python
import numpy as np
data = np.array([
    [101, 5000],
    [102, 7000],
    [103, 9000]
])

print("shape",data.shape)
print("size",data.size)
print("dimensions",data.ndim)
```
