```python
## Four Pillars of OOP  (IMPORTANT POINT TO REMEMBER FOR INTERVIEW)

We'll learn these in detail in upcoming days:
1. Encapsulation - Bundling data and methods together; hiding internal details
2. Inheritance - Creating new classes from existing ones; code reuse
3. Polymorphism - Same interface, different implementations; flexibility
4. Abstraction - Hiding complexity; showing only essential features

## Class and Objects..

Class - Definition: A class is a blueprint or template for creating objects. It defines the structure (attributes) and behavior
(methods) that objects created from the class will have.


Object =  An object is an instance of a class. It's a concrete realization of the class blueprint with specific values for its
attributes.

help(str)

name = str("Lohit")

# from opp perspective you call it instance or object

# str() is a class which is blueprint of the methods you can perform on object
# we got name from class str() which object

name.upper()

## Syntax of class in python

class ClassName:
  ## class body
  # method1
  # method2
  # method3
  pass

## Naming Convention..
# Use PascalCase(capitalize frist letter of each word)
# for example - Car, BankAccount, StudentRecord

## The __init__ constructor..

**What is __init__?**

- Special method (called a "dunder" method - double underscore)
- Automatically called when you create a new object
- Used to initialize the object's attributes with starting values
- Think of it as a "setup" method

name = str("Lohit")

# class str:
  #Capitalize
  #trim
  #strip
  #Startswith

name.upper()

## cl

class ClassName:
  ## basic syntax for __init__ constructor
  def __init__(self, parameter1, parameter2):
    self.attribute1 = parameter1
    self.attribute2 = parameter2

# What is self?
# self represents the instance (object) being created
# It's a reference to the current object
# Always the first parameter in instance methods
# You don't pass it manually - Python does it automatically



## Define the class

# display_info, accelerate, brake

class Car:
  ## Constructor - runs automatically when object is created..
  def __init__(self, brand, model, color):
    self.brand = brand
    self.model = model
    self.color = color
    self.speed = 0  # Default value

  ## Instance methods
  def display_info(self):
    print(f"Car: {self.brand} {self.model}")
    print(f"Color: {self.color}")
    print(f"Current Speed: {self.speed} km/h")

  def accelerate(self, increase):
    self.speed += increase
    print(f"{self.brand} is not going at {self.speed} km/h")

  def brake(self,decrease):
    self.speed -= decrease
    if self.speed < 0:
      self.speed = 0
    print(f"{self.brand} slowed down to {self.speed} km/h")


## Creating Objects(instance) from class Car

name = str("Lohit")


car1 = Car("Toyota", "Camry", "Blue")
car2 = Car("Honda", "Civic", "Red")
car3 = Car("Tesla", "Model 3", "White")


# Object has both data and methods of the class

# Using the objects..

car1.display_info()

car1.accelerate(50)

car1.brake(30)

car1.brake(50)

car1.brand

car1.model

car1.speed

# same for car2

car2.display_info()

car2.accelerate(100)

## Understanding Instance creation
class Dog:
  def __init__(self, name, breed):
    self.name = name
    self.breed = breed
    print(f"A new dog name {name} was created!")
    print("Print this when object is created")

my_dog = Dog("Buddy", "Golder Retriever")

Step-by-step breakdown:

1. Python creates a new empty object
2. Python automatically calls __init__ method
3. self gets the reference to the new object
4. Parameters "Buddy" and "Golden Retriever" are passed
5. Attributes are set: self.name = "Buddy", self.breed = "Golden Retriever"
6. Object is fully initialized and assigned to my_dog

**Key Takeaways:**

1. A class is a blueprint; an object is an instance of that class
2. Use class ClassName: to define a class
3. __init__ is the constructor method that initializes objects
4. self refers to the current instance and must be the first parameter
5. Each object has its own independent copy of instance attributes
6. Methods are functions defined inside a class that operate on objects


**Important Terms:**
- Class: Blueprint/template
- Object/Instance: Concrete realization of a class
- Constructor: __init__ method
- Instance Attribute: Variables specific to each object (e.g., self.name)
- Instance Method: Functions that operate on objects (e.g., def display_info(self))

## Mistake to avoid

1. Forgetting self

# wrong
class Car:
    def __init__(brand, model):  # Missing self
        brand = brand  # This won't work

# corrct
class Car:
    def __init__(self, brand, model):
        self.brand = brand

## Wrong Method Calls

# class Car:
#   #display_info(self)

# car1 = Car()

Car.display_info()    # wrong
car1.display_info()   # correct

## Exercise 1: Movie Class

Create a `Movie` class with:

- **Attributes:**
  - `title` → Name of the movie (string)
  - `director` → Director of the movie (string)
  - `year` → Release year (integer)
  - `rating` → Movie rating (float, e.g., out of 10)

- **Methods:**

  - `display_info()`  
    - Prints or returns all movie details in a readable format  
    - Example output:  
      `"Inception (2010), directed by Christopher Nolan, Rating: 8.8"`

  - `update_rating(new_rating)`  
    - Takes a new rating as input  
    - Updates the movie's `rating` attribute  
    - Optionally validate (e.g., rating should be between 0 and 10)

  - `is_classic()`  
    - Checks if the movie was released before the year 2000  
    - Returns `True` if `year < 2000`, otherwise `False`

## Exercise 2: Challenge – Library Class

Create a `Library` class that can:

- **Store multiple `Book` objects**
  - Use a list (e.g., `self.books = []`) to store all book objects
  - Each book should have attributes like `title`, `author`, etc.

- **Methods:**

  - `add_book(book)`  
    - Takes a `Book` object as input  
    - Adds it to the library collection (`self.books`)  
    - Optionally check for duplicates before adding  

  - `remove_book(title)`  
    - Takes a book title as input  
    - Searches for the book in the collection  
    - Removes it if found  
    - If not found, return or print a message like `"Book not found"`  

  - `display_books()`  
    - Iterates through all books in the library  
    - Prints or returns details of each book  
    - If no books exist, display `"No books available"`  

  - *(Optional Advanced)* `find_book(title)`  
    - Searches for a book by title  
    - Returns the book object or its details if found  

```
